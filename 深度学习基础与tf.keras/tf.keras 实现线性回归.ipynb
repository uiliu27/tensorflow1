### 使用tf.keras 实现简单线性回归

```python
import pandas as pd
```

```python
data=pd.read_csv("./xianxing.txt")
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(data.Education,data.Income)
```

```python
x=data.Education
y=data.Income
```

#### Sequential 顺序模型，它由多个网络层线性堆叠

```python
model=tf.keras.Sequential()
```

#### 1.简单地使用 .add() 来堆叠模型：

```python
model.add(tf.keras.layers.Dense(1,input_shape=(1,)))  #线性层dense的使用技巧
```

```python
model.summary()  
```

#### 2. 使用 .compile() 来配置学习过程：
指定优化器，损失函数等

```python
model.compile(optimizer='adam',loss='mse')   #compile的使用方法
```

#### 3.批量在数据集进行迭代
model.fit()  ,或者手动批次将数据提供给模型 model.train_on_batch(x_batch, y_batch)

```python
history =model.fit(x,y,epochs=5000)
```
