### 序列问题

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
dataset=keras.datasets.imdb
```

```python
max_word=10000
```

```python
(x_train,y_train),(x_test,y_test)=dataset.load_data(num_words=max_word)
```

```python
x_train.shape,y_train.shape
```

```python
 x_train[0]  #用一个整数代替数字
```

### 文本训练成密集向量
文本处理，
<code> keras.preprocessing.sequence.pad_sequences(x_train,300)
    填充到固定长度

```python
[len(x) for x in x_train]
#长度不相同,选哟填充到一个固定的长度
```

```python
x_train=keras.preprocessing.sequence.pad_sequences(x_train,300)
x_test=keras.preprocessing.sequence.pad_sequences(x_test,300)
```

```python
[len(x) for x in x_train]
```

```python
y_train,x_train.shape
```

### 建立模型
文本处理，映射成密集向量

<code>model.add(tf.keras.layers.Embedding(10000,64,input_length=300)) 
    
序列三维数据,将为2维  
    
使用<code>model.add(tf.keras.layers.Flatten()) 展平 ,参数过多
    可以使用 model.add(layers.GlobalAveragePooling1D())   一维全局平均池化

```python
model=keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Embedding(10000,64,input_length=300))
```

```python
model.add(keras.layers.Dropout(0.4))
```

```python
# model.add(layers.Flatten())
model.add(keras.layers.AveragePooling1D())
```

```python
model.add(keras.layers.Dropout(0.4))
```

```python
model.add(layers.Dense(128,activation='relu'))
```

```python
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001),loss='binary_crossentropy',
             metrics=['acc'])
```

```python
history=model.fit(x_train,y_train,epochs=15,batch_size=256,validation_data=(x_test,y_test))
```

```python
history.history.keys()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(history.epoch,history.history['loss'],'r')
plt.plot(history.epoch,history.history['val_loss'],'b')
plt.legend()
```

```python
plt.plot(history.epoch,history.history['acc'],'r')
plt.plot(history.epoch,history.history['val_acc'],'b')
plt.legend()
```
